#1. Análise do faturamento




```python

# Faturamento total por loja
faturamento = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

faturamento
```

```python
for loja, valor in faturamento.items():
    print(f'{loja}: R$ {valor:,.2f}')
```

# 2. Vendas por Categoria


```python
# Agrupar e contar o número de vendas por categoria em cada loja
for nome, df in lojas.items():
    print(f'=== {nome} ===')
    vendas_categoria = df['Categoria do Produto'].value_counts()
    print(vendas_categoria)
    print()
```

```python
# Identificar a categoria mais popular em cada loja
for nome, df in lojas.items():
    categoria_popular = df['Categoria do Produto'].value_counts().idxmax()
    total_vendas = df['Categoria do Produto'].value_counts().max()
    print(f'{nome}: {categoria_popular} ({total_vendas} vendas)')
```

# 3. Média de Avaliação das Lojas


```python
avaliacoes = {
    'Loja 1': loja1['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

for loja, media in avaliacoes.items():
    print(f'{loja}: {media:.2f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
# Produtos mais e menos vendidos em cada loja
for nome, df in lojas.items():
    mais_vendido = df['Produto'].value_counts().idxmax()
    menos_vendido = df['Produto'].value_counts().idxmin()
    print(f'{nome}:')
    print(f' - Produto mais vendido: {mais_vendido}')
    print(f' - Produto menos vendido: {menos_vendido}')
    print()
```

# 5. Frete Médio por Loja

```python
# Cálculo do frete médio em cada loja
frete_medio = {
    'Loja 1': loja1['Frete'].mean(),
    'Loja 2': loja2['Frete'].mean(),
    'Loja 3': loja3['Frete'].mean(),
    'Loja 4': loja4['Frete'].mean()
}

# Exibindo os resultados formatados
for loja, valor in frete_medio.items():
    print(f'{loja}: R$ {valor:.2f}')
```

# **Análise de Desempenho Geográfico**

```python
import matplotlib.pyplot as plt

# Configuração da figura com 4 mapas (um para cada loja)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

# Lista com os DataFrames e nomes das lojas
lojas_lista = [loja1, loja2, loja3, loja4]
titulos = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

# Loop para gerar o mapa de calor de cada loja
for i, ax in enumerate(axes.flat):
    hb = ax.hexbin(
        lojas_lista[i]['lon'],
        lojas_lista[i]['lat'],
        gridsize=30,
        cmap='YlOrRd',
        bins='log'
    )
    ax.set_title(f'Mapa de Calor de Vendas - {titulos[i]}')
    ax.set_xlabel('Longitude')
    ax.set_ylabel('Latitude')
    fig.colorbar(hb, ax=ax, label='Densidade de Vendas')

plt.tight_layout()
plt.show()
```

## 🌍 Padrões Geográficos e Desempenho das Lojas

Ao observar a distribuição geográfica das vendas por meio dos mapas de calor, é possível identificar diferenças relevantes entre as lojas quanto à **abrangência e concentração regional** das transações.

As **Lojas 1 e 3** apresentaram **maior dispersão geográfica**, com pontos de venda distribuídos em diferentes regiões, o que pode estar relacionado ao **maior faturamento total** observado nessas unidades. Essa amplitude de alcance sugere uma **presença comercial mais consolidada** e com maior penetração em diferentes mercados.

Por outro lado, as **Lojas 2 e 4** mostraram **concentração de vendas em áreas específicas**, indicando uma **atuação mais localizada**. A **Loja 4**, em especial, apresentou o **menor faturamento total** entre as quatro, reforçando a hipótese de que **atuações geográficas limitadas** podem influenciar diretamente o **desempenho financeiro**.

Além disso, as regiões de maior densidade de vendas não necessariamente correspondem às áreas com **melhores avaliações dos clientes**, o que sugere que **a localização geográfica não é o único fator determinante** para a satisfação, mas sim uma variável complementar que deve ser considerada junto a fatores como preço, logística e experiência de compra.

💡 **Conclusão:**  
As análises geográficas indicam que **lojas com maior dispersão territorial tendem a ter desempenho superior em faturamento**, enquanto aquelas mais concentradas geograficamente podem enfrentar **limitações de crescimento e alcance de público**. A expansão para novas regiões poderia ser uma estratégia eficaz para equilibrar o desempenho entre as lojas.
